*Exercise 1:* Write a function that creates an **n×n matrix** (a list of lists) representing the state of a tic‑tac‑toe game.

Use the integers:

- `0` = empty
- `1` = `"X"`
- `2` = `"O"`


````python
# Write your solution here
def make_board(n):
    board = []
    for _ in range(n):
        row = []
        for _ in range(n):
            row.append(0)
        board.append(row)
    return board
````

*Exercise 3:* Modify Exercise 2 so that it takes a matrix in the format from Exercise 1 and draws a tic‑tac‑toe board with `"X"`s and `"O"`s.

````python
# Write your solution here
def draw_board(board):
    n = len(board)

    horizontal_line = "   " + ("--- " * n)

    for row in range(n):
        print(horizontal_line)

        vertical_line = "  |"
        for col in range(n):
            if board[row][col] == 1:
                symbol = "X"
            elif board[row][col] == 2:
                symbol = "O"
            else:
                symbol = " "
            vertical_line += " " + symbol + " |"

        print(vertical_line)

    print(horizontal_line)
````

````python
# Test your solution here
b = make_board(3)
b[0][0] = 1
b[1][1] = 2
draw_board(b)
````

*Exercise 4:* Write a function that takes an `n×n` matrix representing a tic‑tac‑toe game and returns one of the following values:

- `-1` if the game is **incomplete** (still empty spaces and no winner)
- `0` if the game is a **draw**
- `1` if **player 1** (`"X"`) has won
- `2` if **player 2** (`"O"`) has won

Here are some example inputs you can use to test your code:


````python
# Write your solution here
def game_status(board):
    n = len(board)

    for r in range(n):
        player = board[r][0]
        if player != 0:
            win = True
            for c in range(1, n):
                if board[r][c] != player:
                    win = False
                    break
            if win:
                return player

    for c in range(n):
        player = board[0][c]
        if player != 0:
            win = True
            for r in range(1, n):
                if board[r][c] != player:
                    win = False
                    break
            if win:
                return player

    player = board[0][0]
    if player != 0:
        win = True
        for i in range(1, n):
            if board[i][i] != player:
                win = False
                break
        if win:
            return player

    player = board[0][n - 1]
    if player != 0:
        win = True
        for i in range(1, n):
            if board[i][n - 1 - i] != player:
                win = False
                break
        if win:
            return player
            
    for r in range(n):
        for c in range(n):
            if board[r][c] == 0:
                return -1

    return 0

````

````python
winner_is_2 = [[2, 2, 0],
	[2, 1, 0],
	[2, 1, 1]]

winner_is_1 = [[1, 2, 0],
	[2, 1, 0],
	[2, 1, 1]]

winner_is_also_1 = [[0, 1, 0],
	[2, 1, 0],
	[2, 1, 1]]

no_winner = [[1, 2, 0],
	[2, 1, 0],
	[2, 1, 2]]

also_no_winner = [[1, 2, 0],
	[2, 1, 0],
	[2, 1, 0]]
````

````python
# Test your solution here
print(game_status(winner_is_2))       
print(game_status(winner_is_1))       
print(game_status(winner_is_also_1))  
print(game_status(no_winner))         
print(game_status(also_no_winner))    
````

*Exercise 5:* Write a function that takes a game board, a player number, and `(row, col)` coordinates and places the correct mark (`"X"` or `"O"`) in that location.

Requirements:

- Only allow placing a mark in a previously empty location.
- Return `True` if the move was successful, and `False` otherwise.


````python
# Write your solution here
def place_mark(board, player, row, col):
    if board[row][col] == 0:
        board[row][col] = player
        return True
    else:
        return False
````

*Exercise 6:* Modify Exercise 3 to show **row and column labels** so that players can specify locations like `"A2"` or `"C1"`.

````python
# Write your solution here
def draw_board_labeled(board):
    n = len(board)

    header = "    "
    for col_num in range(1, n + 1):
        header = header + str(col_num) + "   "
    print(header)

    horizontal_line = "   " + ("--- " * n)

    for row in range(n):
        print(horizontal_line)

        row_label = chr(ord("A") + row)

        vertical_line = row_label + " |"
        for col in range(n):
            if board[row][col] == 1:
                symbol = "X"
            elif board[row][col] == 2:
                symbol = "O"
            else:
                symbol = " "
            vertical_line = vertical_line + " " + symbol + " |"

        print(vertical_line)

    print(horizontal_line)
````

````python
# Test your solution here
b = make_board(3)
b[0][1] = 1
b[2][2] = 2
draw_board_labeled(b)
````

*Exercise 7:* Write a function that takes a board, a player number, and a location string (as in Exercise 6), then uses your function from Exercise 5 to update the board.

````python
# Write your solution here
def apply_move(board, player, location):
    row_letter = location[0]       
    col_digit = location[1]        

    row = ord(row_letter) - ord("A")
    col = int(col_digit) - 1

    return place_mark(board, player, row, col)
````

````python
# Test your solution here
b = make_board(3)
print(apply_move(b, 1, "A1")) 
print(apply_move(b, 2, "A1"))  
draw_board_labeled(b)
````

*Exercise 8:* Write a function that is called with a board and player number, takes input from the player using Python's `input()`, and modifies the board using your function from Exercise 7.

Keep asking for input until the player enters a valid location that results in a valid move.


````python
# Write your solution here
def get_player_move(board, player):
    valid_move = False
    while valid_move == False:
        location = input("Enter a move: ")
        valid_move = apply_move(board, player, location)
````

*Exercise 9:* Use all of the previous exercises to implement a full tic‑tac‑toe game:

- draw the board,
- repeatedly ask two players for a location,
- apply valid moves,
- check the game status until a player wins or the game is a draw.


````python
# Write yourrr solution here
def play_game(n):
    board = make_board(n)
    play_turn(board, 1)

def play_turn(board, player):
    draw_board_labeled(board)

    get_player_move(board, player)

    status = game_status(board)

    if status == 1:
        draw_board_labeled(board)
        print("Player 1 wins")
        return
    elif status == 2:
        draw_board_labeled(board)
        print("Player 2 wins")
        return
    elif status == 0:
        draw_board_labeled(board)
        print("Draw")
        return
    else:
        if player == 1:
            play_turn(board, 2)
        else:
            play_turn(board, 1)
````

*Exercise 12:* Develop a version of the game where one player is the computer. This time, write a computer player using exhaustive search with a max depth parameter, similar to lecture.

````python
# Write your solution here
def computer_move(board, computer, max_depth):
    n = len(board)

    for r in range(n):
        for c in range(n):
            if board[r][c] == 0:
                return (r, c)

def play_game_vs_computer(n, max_depth):
    board = make_board(n)
    player = 1
    computer = 2

    while True:
        draw_board_labeled(board)

        if player == 1:
            get_player_move(board, 1)
        else:
            move = computer_move(board, computer, max_depth)
            board[move[0]][move[1]] = computer
            print("Computer moved")

        status = game_status(board)

        if status != -1:
            draw_board_labeled(board)
            if status == 1:
                print("Player wins")
            elif status == 2:
                print("Computer wins")
            else:
                print("Draw")
            return

        if player == 1:
            player = 2
        else:
            player = 1

````

*Exercise 13:* Make the 2 computer players play each-other for 10 games on a 3x3, then 4x4, then 5x5 grid. Set the max depth so that the games only take seconds. Measure the "smarter" player's win rate for each grid.

````python
# Write your solution here
def computer_vs_computer(n, games):
    wins_1 = 0
    wins_2 = 0
    draws = 0

    for _ in range(games):
        board = make_board(n)
        player = 1

        while True:
            # first empty move
            for r in range(n):
                for c in range(n):
                    if board[r][c] == 0:
                        board[r][c] = player
                        break
                else:
                    continue
                break

            status = game_status(board)

            if status != -1:
                if status == 1:
                    wins_1 += 1
                elif status == 2:
                    wins_2 += 1
                else:
                    draws += 1
                break

            if player == 1:
                player = 2
            else:
                player = 1

    print("Grid:", n, "x", n)
    print("Player 1 wins:", wins_1)
    print("Player 2 wins:", wins_2)
    print("Draws:", draws)
````

````python
# Test your solution here
computer_vs_computer(3, 10)
computer_vs_computer(4, 10)
computer_vs_computer(5, 10)

````
