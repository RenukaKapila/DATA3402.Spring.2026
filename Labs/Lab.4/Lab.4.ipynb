1. Write a "counter" class that can be incremented up to a specified maximum value, will print an error if an attempt is made to increment beyond that value, and allows reseting the counter. 

```python
class Counter:

    def __init__(self, max_value):
        
        # store the maximum allowed value
        self.max_value = max_value
        
        # counter starts at 0
        self.value = 0


    def increment(self):
        
        # check if counter already reached maximum
        if self.value >= self.max_value:
            print("Error: counter at maximum value")
        
        else:
            # increase counter by 1
            self.value = self.value + 1


    def reset(self):
        
        # reset counter back to zero
        self.value = 0


# create counter with maximum value 3
c = Counter(3)

print(c.value)

c.increment()
print(c.value)

c.increment()
print(c.value)

c.increment()
print(c.value)

# this should produce an error
c.increment()

# reset the counter
c.reset()
print(c.value)
```

3. Implement a class to represent a rectangle, holding the length, width, and $x$ and $y$ coordinates of a corner of the object. Implement functions that compute the area and perimeter of the rectangle. Make all data members private and privide accessors to retrieve values of data members. 

```python
class Rectangle:

    def __init__(self, length, width, x, y):

        # store the rectangle dimensions
        # these represent the size of the rectangle
        self.__length = length
        self.__width = width

        # store the position of one corner of the rectangle
        # this is the bottom-left corner coordinate
        self.__x = x
        self.__y = y


    def get_length(self):

        # accessor function that returns the rectangle length
        return self.__length


    def get_width(self):

        # accessor function that returns the rectangle width
        return self.__width


    def get_x(self):

        # accessor function that returns the x coordinate
        return self.__x


    def get_y(self):

        # accessor function that returns the y coordinate
        return self.__y


    def area(self):

        # area of a rectangle = length * width
        return self.__length * self.__width


    def perimeter(self):

        # perimeter = sum of all sides
        # rectangle has two lengths and two widths
        return 2 * (self.__length + self.__width)

# create a rectangle object
r = Rectangle(10, 5, 2, 3)

# test the accessor functions
print("Length:", r.get_length())
print("Width:", r.get_width())
print("X position:", r.get_x())
print("Y position:", r.get_y())

# test area calculation
print("Area:", r.area())

# test perimeter calculation
print("Perimeter:", r.perimeter())
```

4. Implement a class to represent a circle, holding the radius and $x$ and $y$ coordinates of center of the object. Implement functions that compute the area and perimeter of the rectangle. Make all data members private and privide accessors to retrieve values of data members. 

```python
# define a constant value for pi
PI = 3.14


class Circle:

    def __init__(self, radius, x, y):

        # store the radius of the circle
        self.__radius = radius

        # store the center position of the circle
        self.__x = x
        self.__y = y


    def get_radius(self):

        # accessor that returns the radius
        return self.__radius


    def get_x(self):

        # accessor that returns the x coordinate of the center
        return self.__x


    def get_y(self):

        # accessor that returns the y coordinate of the center
        return self.__y


    def area(self):

        # area formula for a circle
        # area = pi * r^2
        return PI * self.__radius * self.__radius


    def perimeter(self):

        # perimeter (circumference) formula
        # perimeter = 2 * pi * r
        return 2 * PI * self.__radius


# create a circle object
c = Circle(4, 1, 2)

# test the accessors
print("Radius:", c.get_radius())
print("Center X:", c.get_x())
print("Center Y:", c.get_y())

# test calculations
print("Area:", c.area())
print("Perimeter:", c.perimeter())
```

```python
class Triangle(Shape):

    def __init__(self, base, height, x, y):

        # store triangle dimensions
        self.__base = base
        self.__height = height

        # store triangle position
        self.__x = x
        self.__y = y


    def get_base(self):
        return self.__base


    def get_height(self):
        return self.__height


    def get_x(self):
        return self.__x


    def get_y(self):
        return self.__y


    def area(self):

        # area = 1/2 * base * height
        return 0.5 * self.__base * self.__height


    def perimeter(self):

        # compute hypotenuse
        hyp = self.__sqrt((self.__base * self.__base) + (self.__height * self.__height))

        return self.__base + self.__height + hyp


    def __sqrt(self, value):

        # Newton method square root
        guess = value

        for i in range(20):
            guess = 0.5 * (guess + value / guess)

        return guess


t = Triangle(3,4,0,0)

print("Base:", t.get_base())
print("Height:", t.get_height())
print("Area:", t.area())
print("Perimeter:", t.perimeter())
```

8. Add a function to the object classes, including the base, that tests if a given set of $x$ and $y$ coordinates are inside of the object. You'll have to think through how to determine if a set of coordinates are inside an object for each object type.

```python
class Shape:

    def get_points(self):
        return []

    def contains(self, x, y):
        # not implemented for generic Shape
        return False

class Rectangle(Shape):

    def __init__(self, length, width, x, y):
        self.__length = length
        self.__width = width
        self.__x = x
        self.__y = y

    def contains(self, x, y):

        left = self.__x
        right = self.__x + self.__length

        bottom = self.__y
        top = self.__y + self.__width

        if x >= left and x <= right and y >= bottom and y <= top:
            return True
        else:
            return False

class Circle(Shape):

    def __init__(self, radius, x, y):
        self.__radius = radius
        self.__x = x
        self.__y = y

    def contains(self, x, y):

        cx = self.__x
        cy = self.__y
        r = self.__radius

        dx = x - cx
        dy = y - cy

        if (dx * dx) + (dy * dy) <= (r * r):
            return True
        else:
            return False

class Triangle(Shape):

    def __init__(self, base, height, x, y):
        self.__base = base
        self.__height = height
        self.__x = x
        self.__y = y

    def contains(self, x, y):

        x0 = self.__x
        y0 = self.__y
        b = self.__base
        h = self.__height

        # quick bounding box test
        if x < x0 or x > x0 + b or y < y0 or y > y0 + h:
            return False

        dx = x - x0
        dy = y - y0

        left_side = (dx * h) + (dy * b)
        right_side = b * h

        if left_side <= right_side:
            return True
        else:
            return False


r = Rectangle(10, 5, 0, 0)
c = Circle(4, 2, 3)
t = Triangle(3, 4, 0, 0)

print(r.contains(5, 2))   
print(r.contains(20, 2))  
```

9. Add a function in the base class of the object classes that returns true/false testing that the object overlaps with another object.

```python
class Shape:
    def get_points(self):
        return []

    def contains(self, x, y):
        return False

    def overlaps(self, other):

        for (x, y) in self.get_points():

            if other.contains(x, y):
                return True

        return False

class Rectangle(Shape):

    def __init__(self, length, width, x, y):
        self.__length = length
        self.__width  = width
        self.__x = x
        self.__y = y


    def get_points(self):

        x = self.__x
        y = self.__y
        L = self.__length
        W = self.__width

        points = []
        points.append((x, y))
        points.append((x + L, y))
        points.append((x + L, y + W))
        points.append((x, y + W))

        return points


    def contains(self, x, y):

        left   = self.__x
        right  = self.__x + self.__length
        bottom = self.__y
        top    = self.__y + self.__width

        return (left <= x <= right) and (bottom <= y <= top)


class Circle(Shape):

    def __init__(self, radius, x, y):
        self.__radius = radius
        self.__x = x
        self.__y = y


    def get_points(self):

        cx = self.__x
        cy = self.__y
        r  = self.__radius

        points = []
        points.append((cx + r, cy))
        points.append((cx - r, cy))
        points.append((cx, cy + r))
        points.append((cx, cy - r))

        return points


    def contains(self, x, y):

        dx = x - self.__x
        dy = y - self.__y

        return (dx*dx + dy*dy) <= (self.__radius * self.__radius)

class Triangle(Shape):

    def __init__(self, base, height, x, y):
        self.__base = base
        self.__height = height
        self.__x = x
        self.__y = y


    def get_points(self):

        x = self.__x
        y = self.__y
        b = self.__base
        h = self.__height

        points = []
        points.append((x, y))
        points.append((x + b, y))
        points.append((x, y + h))

        return points


    def contains(self, x, y):

        x0 = self.__x
        y0 = self.__y
        b  = self.__base
        h  = self.__height

        if x < x0 or y < y0:
            return False
        if x > x0 + b or y > y0 + h:
            return False

        return ((x - x0) / b) + ((y - y0) / h) <= 1



r = Rectangle(10,5,0,0)
c = Circle(4,2,3)
t = Triangle(3,4,0,0)

print(r.overlaps(c))
print(c.overlaps(r))
print(r.overlaps(t))

```

```python
class foo:
    def __init__(self,a,b=None):
        self.a=a
        self.b=b
        
    def __repr__(self):
        return "foo("+repr(self.a)+","+repr(self.b)+")"
    
    def save(self,filename):
        f=open(filename,"w")
        f.write(self.__repr__())
        f.close()
        
   
def foo_loader(filename):
    f=open(filename,"r")
    tmp=eval(f.read())
    f.close()
    return tmp
```
